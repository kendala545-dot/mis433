## Part 2: Data Summarization

```python
# Total "Gross" of all top 1000 movies
total_gross = imdb['Gross'].sum()
print(f"Total Gross of all top 1000 movies: ${total_gross:,.2f}")
```

```python
# Average "No_of_Votes" of all movies
average_votes = imdb['No_of_Votes'].mean()
print(f"Average number of votes for all movies: {average_votes:,.2f}")
```

```python
# Count movies in each decade (e.g., ..., 1980, 1990, 2000, 2010, 2020)
# Sort decades by the number of movies in descending order
decade_counts = imdb['Decade'].value_counts().sort_index()
display(decade_counts)
```

```python
# Count movies by different directors.
# Show the top 10 directors with the most movies in this list
director_counts = imdb['Director'].value_counts()
display(director_counts.head(10))
```

```python
# For each director, calculate the number of movies, average IMDB_Rating, and total gross.
# Ranked in descending order of total gross
# Show the top 10 directors with the most gross
director_summary = imdb.groupby('Director').agg(
    num_movies=('Series_Title', 'count'),
    average_imdb_rating=('IMDB_Rating', 'mean'),
    total_gross=('Gross', 'sum')
).sort_values(by='total_gross', ascending=False)

display(director_summary.head(10))
```

```python
# Group movies by decade and director
# In each group (i.e., for each decade and each director),
# calculate the number of movies and average IMDB rating.
# Sort in descending order of movie count
decade_director_summary = imdb.groupby(['Decade', 'Director']).agg(
    num_movies=('Series_Title', 'count'),
    average_imdb_rating=('IMDB_Rating', 'mean')
).sort_values(by='num_movies', ascending=False)

display(decade_director_summary.head(10))
```

```python
# Find the top 3 highest rated movies of each year since 2010.
# Show their released year, ranking, title, and IMDB_Rating.
# Sort them in descending order of year and ascending order of ranking

# Filter for movies released in or after 2010
movies_since_2010 = imdb[imdb['Released_Year'] >= 2010]

# Group by 'Released_Year' and select the top 3 by 'IMDB_Rating'
top_movies_per_year = movies_since_2010.groupby('Released_Year').apply(
    lambda x: x.nlargest(3, 'IMDB_Rating')
).reset_index(drop=True)

# Select desired columns and sort
result = top_movies_per_year[['Released_Year', 'Series_Title', 'IMDB_Rating']].sort_values(
    by=['Released_Year', 'IMDB_Rating'],
    ascending=[False, False]
)

result
```

```python
# Find all directors whose movies appeared in at least five different decades.
# Your result should include: director, decade, and the number of movies in the decade.

# Group by Director and Decade to count movies per decade for each director
director_decade_counts = imdb.groupby(['Director', 'Decade']).size().reset_index(name='num_movies')

# Count the number of unique decades each director has movies in
decade_diversity = director_decade_counts.groupby('Director')['Decade'].nunique().reset_index(name='num_decades')

# Filter for directors who have movies in at least five different decades
multi_decade_directors = decade_diversity[decade_diversity['num_decades'] >= 5]

# Filter the original director_decade_counts to include only these directors
result = director_decade_counts[director_decade_counts['Director'].isin(multi_decade_directors['Director'])]

# Sort for better readability (optional)
result = result.sort_values(by=['Director', 'Decade'])

display(result)
```
