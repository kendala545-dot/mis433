## Question 1. Selecting rows

From the 'flights' dataframe, find all flights that satisfy the following certain conditions:

```python
# Had an arrival delay of two or more hours
delayed_flights = flights[flights['arr_delay'] >= 120]
display(delayed_flights)
```

```python
# Flew to Houston (IAH or HOU)
houston_flights = flights[flights['dest'].isin(['IAH', 'HOU'])]
display(houston_flights)
```

```python
# Were operated by United (UA), American (AA), or Delta (DL)
major_carrier_flights = flights[flights['carrier'].isin(['UA', 'AA', 'DL'])]
display(major_carrier_flights)
```

```python
# Departed in July, August, and September
summer_flights = flights[flights['month'].isin([7, 8, 9])]
display(summer_flights)
```

```python
# Arrived more than two hours late, but didn’t leave late
late_arrival_early_departure_flights = flights[(flights['arr_delay'] >= 120) & (flights['dep_delay'] <= 0)]
display(late_arrival_early_departure_flights)
```

```python
# Were delayed by at least an hour, but made up over 30 minutes in flight
made_up_time_flights = flights[(flights['dep_delay'] >= 60) & ((flights['dep_delay'] - flights['arr_delay']) > 30)]
display(made_up_time_flights)
```

```python
# Departed between midnight and 6am (inclusive)
early_morning_flights = flights[(flights['dep_time'] >= 0) & (flights['dep_time'] <= 600)]
display(early_morning_flights)
```

```python
# How many flights have a missing dep_time?
missing_dep_time_count = flights['dep_time'].isnull().sum()
print(f"Number of flights with missing dep_time: {missing_dep_time_count}")
```

## Question 4. Adding new columns

Currently dep_time and sched_dep_time are convenient to look at, but hard to compute with because they’re not really continuous numbers.

For example, 759 means 7:59 and 801 means 8:01. Their difference is not 42 but 2 minutes.

```python
# Convert them to a more convenient representation of number of minutes since midnight (0).
flights['dep_time_min'] = flights['dep_time'].apply(lambda x: int(x // 100) * 60 + int(x % 100) if pd.notnull(x) else x)
flights['sched_dep_time_min'] = flights['sched_dep_time'].apply(lambda x: int(x // 100) * 60 + int(x % 100) if pd.notnull(x) else x)

display(flights[['dep_time', 'dep_time_min', 'sched_dep_time', 'sched_dep_time_min']].head())
```

```python
# Create a new column of arr_time - dep_time.
flights['arr_time_min'] = flights['arr_time'].apply(lambda x: int(x // 100) * 60 + int(x % 100) if pd.notnull(x) else x)
flights['calculated_flight_duration'] = flights['arr_time_min'] - flights['dep_time_min']

# Compare this column with air_time.
display(flights[['dep_time_min', 'arr_time_min', 'calculated_flight_duration', 'air_time']].head())
```

```python
# Compare dep_time, sched_dep_time, and dep_delay. How would you expect those three numbers to be related?
# Dep_delay is the difference between the dep_time and the sched_dep_time

# Try creating a column to calculate dep_delay from dep_time and sched_dep_time (and/or other columns if necessary).
flights['calculated_dep_delay'] = flights['dep_time_min'] - flights['sched_dep_time_min']

# Test your results.
display(flights[['dep_time', 'dep_time_min', 'sched_dep_time', 'sched_dep_time_min', 'dep_delay', 'calculated_dep_delay']].head())
```

```python
# Sort all AA flights to find the top 10 fastest (highest speed) flights.
# Display the following: year,month,day,carrier,flight,orig,dest,distance,air_time,speed (miles per hour)
aa_flights = flights[flights['carrier'] == 'AA'].copy()
aa_flights['speed'] = aa_flights['distance'] / (aa_flights['air_time'] / 60)

fastest_aa_flights = aa_flights.sort_values(by='speed', ascending=False)

display(fastest_aa_flights[['year', 'month', 'day', 'carrier', 'flight', 'origin', 'dest', 'distance', 'air_time', 'speed']].head(10))
```
