## DataFrame with columns

- year,month,day
        Date of departure    
- dep_time,arr_time
        Actual departure and arrival times (format HHMM or HMM), local tz.
- sched_dep_time,sched_arr_time
        Scheduled departure and arrival times (format HHMM or HMM), local tz.    
- dep_delay,arr_delay
        Departure and arrival delays, in minutes. Negative times represent early departures/arrivals.
- hour,minute
        Time of scheduled departure broken into hour and minutes.
- carrier
        Two letter carrier abbreviation. See airlines() to get name
- tailnum
        Plane tail number
- flight
        Flight number
- origin,dest
        Origin and destination. See airports() for additional metadata.
- air_time
        Amount of time spent in the air, in minutes
- distance
        Distance between airports, in miles
- time_hour
        Scheduled date and hour of the flight as a date. Along with origin, can be used to join flights data to weather data.

## Exercises

Write Python scripts to answer the following questions.

```python
# How many distinct airplanes are there in this dataset?
num_distinct_airplanes = flights['tailnum'].nunique()
print(f"Number of distinct airplanes: {num_distinct_airplanes}")
```

```python
# Which hours of the day are the busiest?
# List the top 5 hours with the most flights.
busiest_hours = flights['hour'].value_counts().head(5)
print("Top 5 busiest hours with the most flights:")
busiest_hours
```

```python
# Find all flights (identified by 'carrier' and 'flight')
# that always depart at least 60 minutes late.
df1 = flights.groupby(['carrier', 'flight']).agg({'dep_delay':'min'})
df1[df1.dep_delay>=60]
```

```python
# How many flights always depart at least 60 minutes late?

df1 = (
    flights
    .groupby(['carrier','flight'])
    .agg({'dep_delay':"min"})
    .query('dep_delay>=60')
).shape[0]
print(f"Number of flights that always depart at least 60 minutes late: {df1}")
```

```python
# Find all United Airlines (UA) flights (identified by 'carrier' and 'flight')
# that depart (at least) 60 minutes late.
(
    flights[flights.carrier=="UA"]
    .groupby(['carrier','flight'])
    .agg({'dep_delay':"min"})
    .query('dep_delay>=60')

).shape[0]
```

```python
# How many United Airlines (UA) flights depart (at least) 60 minutes late?
(
    flights[flights.carrier=="UA"]
    .groupby(['carrier','flight'])
    .agg({'dep_delay':"min"})
    .query('dep_delay>=60')

).shape[0]
```

```python
# Which flights that always arrived on time in December?
(
flights[flights.month==12]
.groupby(['carrier','flight'])
.agg({'arr_delay':'max'})
.query('arr_delay<=0')
)
```

```python
# How many flights always arrived on time in December?
df1 = (
flights[flights.month==12]
.groupby(['carrier','flight'])
.agg({'arr_delay':'max'})
.query('arr_delay<=0')
)
print(f"Number of flights always on time in December: {df1.shape[0]}")
```

```python
# Create a dataframe 'cancelled' for all cancelled flights.
# Assumption: flights that are not cancelled should have values for dep_time.
cancelled = flights[flights['dep_time'].isnull()]
display(cancelled.head())
```

```python
# Look at the number of cancelled flights per day.
# Find the top 10 days with the most cancelled flights.
cancelled[['year','month', 'day']].value_counts()
```

```python
# Which carrier has the worst average arrival delay?

worst_carrier = flights.groupby('carrier').agg({'arr_delay':'mean'}).sort_values('arr_delay', ascending=False).head(1)
print(f"The carrier with the worst average arrival delay is {worst_carrier.index[0]} with an average delay of {worst_carrier.iloc[0,0]:.2f} minutes.")
```

```python
# Which plane (tailnum) has the worst on-time record?
df1 = flights.groupby('tailnum').agg({'arr_delay':['mean','min','max']})
df1.columns = ['ad_mean', 'ad_min','ad_max'] # Renaming columns flattens the MultiIndex
# Filter out rows where ad_mean is NaN before sorting
df1 = df1[df1['ad_mean'].notna()]
# Sort by the mean arrival delay in descending order to find the worst record
worst_plane = df1.sort_values(by='ad_mean', ascending=False).head(1)
print(f"The plane with the worst on-time record (highest average arrival delay) is {worst_plane.index[0]} with an average delay of {worst_plane['ad_mean'].iloc[0]:.2f} minutes.")
```

```python
# What time (hour) of day should you fly if you want to avoid delays as much as possible?
avg_delays_per_hour = flights.groupby('hour').agg(mean_dep_delay=('dep_delay', 'mean'))

best_hour = avg_delays_per_hour['mean_dep_delay'].idxmin()
min_delay = avg_delays_per_hour['mean_dep_delay'].min()

print(f"The best time to fly to avoid delays as much as possible is at hour {int(best_hour)} with an average departure delay of {min_delay:.2f} minutes.")
```

```python
# For each destination, compute the total minutes of arrival delay.
flights.groupby('dest').agg({'arr_delay':'sum'})
```

```python
# Find the worst days in terms of average arrival delay greater than 1 hour.
flights.groupby(['year','month','day']).agg(mean_arr_delay=('arr_delay','mean')).query('mean_arr_delay>60').sort_values('mean_arr_delay', ascending=False).tail(10)
```

```python
# Find the number of carriers between each pair of origin and destination.
flights.groupby(['origin', 'dest']).agg({'carrier':'nunique'}).reset_index()
```
